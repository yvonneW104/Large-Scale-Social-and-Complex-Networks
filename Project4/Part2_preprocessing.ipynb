```python
import re
import csv
```

```python
movie_name_id = {}
movie_cast_ID = {}
movie_set = set()
target = re.compile(r"\s*\([^()]*\)\s*")
with open('movie_dict.txt', 'r', encoding='latin-1') as f:
    for line in f:
        items = line.split("||")
        movie_id, movie_name, casts = items[0], target.sub('', items[1]), items[2]
        cast_list = casts.strip().split("\t\t")
        movie_name_id[movie_name] = movie_id
        movie_cast_ID[movie_id] = cast_list
```

```python
with open('movie_name_id.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(['movie_name', 'movie_id'])
    for movie_name in movie_name_id.keys():
        spamwriter.writerow([movie_name, movie_name_id[movie_name]])
```

```python
with open('movie_cast_id.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(['movie_id', 'cast_id'])
    for movie_id in movie_cast_ID.keys():
        for cast_id in movie_cast_ID[movie_id]:
            spamwriter.writerow([movie_id, cast_id])
```

```python
genre_set = set()
with open("movie_genre.txt", 'r', encoding='latin-1') as f:
    for line in f:
        genre_set.add(line.strip().split("\t\t")[1])
        
genre_id = {genre_name: genre_id for genre_id, genre_name in enumerate(genre_set)}

with open('genre_id.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(['genre_name', 'genre_id'])
    for fromID in genre_id.keys():
        spamwriter.writerow([fromID, genre_id[fromID]])
```

```python
movie_genre_ID = {}
target = re.compile(r"\s*\([^()]*\)\s*")
with open("movie_genre.txt", 'r', encoding='latin-1') as f:
    for line in f:
        items = line.strip().split("\t\t")
        movie_name, genre_name = target.sub('', items[0]), items[1]
        if movie_name in movie_name_id.keys():
            movie_genre_ID[movie_name_id[movie_name]] = genre_id[genre_name]
```

```python
with open('movie_genre_ID.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(['movie_id', 'genre_id'])
    for movie_id in movie_genre_ID.keys():
        spamwriter.writerow([movie_id, movie_genre_ID[movie_id]])
```

```python
movie_rate_ID = {}
target = re.compile(r"\s*\([^()]*\)\s*")
with open("movie_rating.txt", 'r', encoding='latin-1') as f:
    for line in f:
        items = line.strip().split("\t\t")
        movie_name, rating = target.sub('', items[0]), items[1]
        if movie_name in movie_name_id.keys():
            movie_rate_ID[movie_name_id[movie_name]] = rating
        
with open('movie_rating_id.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(['movie_id', 'rating'])
    for movie_id in movie_rate_ID.keys():
        spamwriter.writerow([movie_id, movie_rate_ID[movie_id]])
```

```python
community = {}
num = 1
with open('community.txt', 'r', encoding='latin-1') as f:
    for line in f:
        items = line.strip().split(" ")
        community[num] = items
        num = num + 1
```

```python
with open('community.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(['group', 'member'])
    for group in community.keys():
        for member in community[group]:
            spamwriter.writerow([group, member])
```

```python
counter
```

```python
sum_ = sum(counter.values())
with open('genre_freq.csv', 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(["genre", "freq"])
    for i in counter.keys():
        spamwriter.writerow([i, counter[i]/sum_])
```
